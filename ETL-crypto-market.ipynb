```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import numpy as np
```

```python
Driver='ODBC Driver 17 for SQL Server'
Server='DEEPRAI\SQLEXPRESS'
database='new'
```

```python
url=f'mssql://@{Server}/{database}?driver={Driver}'
```

```python
engine=create_engine(url)
```

```python
try:
    con=engine.connect()
    print('db Connected')
    print('Connection Object is:{}'.format(con))
except Exception as e:
    print(e)
```

```python
df=pd.read_csv("crypto-markets.csv")
```

```python
df.head()
```

```python
assetsCode = ['BTC','ETH','XRP','LTC']
```

```python
df['open'] = df[['open', 'asset']].apply(lambda x: (float(x[0]) * 0.80) if x[1] in assetsCode else np.nan, axis=1)
df['close'] = df[['close', 'asset']].apply(lambda x: (float(x[0]) * 0.80) if x[1] in assetsCode else np.nan, axis=1)
df['high'] = df[['high', 'asset']].apply(lambda x: (float(x[0]) * 0.80) if x[1] in assetsCode else np.nan, axis=1)
df['low'] = df[['low', 'asset']].apply(lambda x: (float(x[0]) * 0.80) if x[1] in assetsCode else np.nan, axis=1)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True ,inplace=True)
df.head()
```

```python
df.drop(labels=['slug', 'ranknow', 'volume', 'market', 'close_ratio', 'spread'], inplace=True, axis=1)
df.head()
```

```python
try:
    df.to_sql("HELLO.crypto",con)
    print("sent to SQL Server")
except Exception as e:
    print(e)
```
