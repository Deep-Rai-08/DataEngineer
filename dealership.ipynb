```python
def extract_csv(file):
    df=pd.read_csv(file)
    return df
def extract_xml(file):
    df=pd.read_xml(file)
    return df
def extract_json(file):
    df=pd.read_json(file)
    return df
```

```python
def extract():
    extracted_data=pd.DataFrame(columns=['car_model','year_of_manufacture','price', 'fuel'])
    try:
        for csvfile in glob.glob("dealership_data/*.csv"):
            df=extract_csv(csvfile)
            extracted_data=pd.concat([extracted_data,df])
        for csvfile in glob.glob("dealership_data/*.xml"):
            df=extract_xml(csvfile)
            extracted_data=pd.concat([extracted_data,df])
        return extracted_data
    except Exception as e:
        print(e)
    
```

```python
def transform(df):
    df['price']=round(df.price,2)
    return df
```

```python
def load(data_to_load):
    data_to_load.to_csv('dealership.csv')
```

```python
def log(message):
    timestamp_format = '%H:%M:%S-%h-%d-%Y'
    #Hour-Minute-Second-MonthName-Day-Year
    now = datetime.now() # get current timestamp
    timestamp = now.strftime(timestamp_format)
    with open("car_dealership_logfile.txt","a") as f: f.write("\n"+timestamp + ',' + message)
```

```python
log("Extracting the data from different file and storing in single dataframe")
fetch=extract()
```

```python
log("Transforming the column 'Price' and roundoff to two decimal points")
transform(fetch)
```

```python
log("Loading the data into a csv file")
load(fetch)
```
